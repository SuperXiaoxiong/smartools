```python
def get_file(tar_dir, file_dir):
    data ={"path" : tar_dir, "password":""}
    file_list = []
    
    try:
        res = requests.post("http://pan.wgpsec.org/api/path",json = data, proxies=proxies)
        print(data)

        res_json_data = res.json().get('data', [])



        for item in res_json_data:
            if item.get('type') == 'file':
                file_list.append(item.get('dir', '') + item.get('name', ''))
            else:
                file_list.extend(get_file(item.get('dir', '') + item.get('name', ''), []))

        print("in dir{}".format(tar_dir))
    except Exception as e:
        print(e)
    return file_list
    
```

```python
import os
def write_to_file(file_list):
    base_url = "http://pan.wgpsec.org/d/"
    base_file = '/Users/admin/work/未消化资料/'
    for item in file_list:
        
        try:
            target_url = base_url + item
            res = requests.get(target_url )
            target_file = base_file + item
            
            (dir_name, file_name) = os.path.split(target_file)
            if not os.path.exists(dir_name):
                os.makedirs(dir_name)
            with open(target_file, 'wb') as f:
                f.write(res.content)    
        except Exception as e:
            print("error in download {}, error is {}".format(item, e))
            
```

```python
write_to_file(file_list)

```
